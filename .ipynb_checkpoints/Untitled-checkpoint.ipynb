```python
df_last["분양가격(㎡)"].astype(int)
# object타입이기 때문에 연산불가능
# int로 변환 해야함 but" "(공백)이 있기때문에 String으로 인식해서 변환불가능
# pandas의 to_numeric매서드를 통해 변환
```

```python
type(pd.np.nan)
```

```python
df_last["분양가격"]=pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
df_last["분양가격"].head(1)
# errors='coerce' 타입은 오류가 났을때 강제로 변환하게함
```

```python
df_last["평당분양가격"]=df_last["분양가격"] *3.3
df_last.head(1)
#가격을 비교하기 위해서 평당분양가격(3.3㎡) 으로 바꿈 -> 기존데이터와 비교해보기 위해서임
```

```python
df_last.info()
```

```python
df_last["분양가격(㎡)"].describe()
# 변경전 칼럼인 "분양가격(㎡)"(object) 요약
# unique = 중복되지 않은 값
# top = 가장 빈번하게 등장하는 문자
# freq = 가장 빈번하게 등장하는 문자가 몇번 등장하는지
```

```python
df_last["분양가격"].describe()
# 변경후 칼럼인 "분양가격"(float) 요약
# std = 표준편차
# mean = 평균값
# count가 달라지는 이유 "분양가격(㎡)"에는 공백 데이터가 있지만 "분양가격"에서는 결측치(손실데이터)로 변경이되었기 때문에 달라짐
```

```python
df_last["규모구분"].unique()
```

```python
df_last["전용면적"]=df_last["규모구분"].str.replace("전용면적","")
df_last["전용면적"]=df_last["전용면적"].str.replace("초과","~")
df_last["전용면적"]=df_last["전용면적"].str.replace("이하", "")
df_last["전용면적"]=df_last["전용면적"].str.replace(" ","").str.strip()
df_last["전용면적"]
#replace는 text가 완전히 같아야지 바꿔줌 -> str에있는 replace로 써야함
#str.strip 앞뒤 공백을 제거한다
```

```python
df_last.info()
```

```python
df_last=df_last.drop(["규모구분","분양가격(㎡)"],axis=1)
df_last
# 1 = col 0 = row
# 0으로하면 row에는 저 배열값이 없기때문에 오류가남
```
